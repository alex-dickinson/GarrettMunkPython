## Example 1: Vertical spectra

Plot vertical wavenumber spectra for the GM75, GM76 and GK91 models

Compute spectra directly and also by integration over kx-kz spectra for comparison

```python
import numpy as np
import matplotlib.pyplot as plt

import gm_toolbox

# TODO add gsw
# import gsw
```

### Set up wavenumber ranges and local parameters

```python

### All calculations are done in terms of kz and kx
logkz_min = -4.
logkz_max = 1.
Nkz = 100
kz = np.logspace(logkz_min, logkz_max, Nkz)

### Set local parameters
N = 5.24e-3
# f_rads = gsw.f(30.)
f = 7.2921e-5
```

### Compute GM75 kz-spectrum using direct expressions

```python
### Set parameters for GM75:
E = 6.3e-5
b = 1300.
jstar = 6.
N0 = 5.24e-3

gm_form = "gm75"
rolloff_option = "no"
A_gm75_unsat = gm_toolbox.vertical_spectrum_Akz(kz, jstar, N, N0, b, gm_form, rolloff_option)
rolloff_option = "yes"
A_gm75_sat = gm_toolbox.vertical_spectrum_Akz(kz, jstar, N, N0, b, gm_form, rolloff_option)

plt.loglog(kz, A_gm75_unsat, label="GM75 unsat")
plt.loglog(kz, A_gm75_sat, label="GM75 sat")

plt.legend()
plt.show()
```

### Compute GM76 kz-spectrum using direct expressions

```python
### Set parameters for GM76:
E = 6.3e-5
b = 1300.
jstar = 3.
N0 = 5.24e-3

gm_form = "gm76"
rolloff_option = "no"
A_gm76_unsat = gm_toolbox.vertical_spectrum_Akz(kz, jstar, N, N0, b, gm_form, rolloff_option)
rolloff_option = "yes"
A_gm76_sat = gm_toolbox.vertical_spectrum_Akz(kz, jstar, N, N0, b, gm_form, rolloff_option)

plt.loglog(kz, A_gm76_unsat, label="GM76 unsat")
plt.loglog(kz, A_gm76_sat, label="GM76 sat")

plt.legend()
plt.show()
```

### Compute GK91 kz-spectrum using direct expressions

```python
### Set parameters for GK91:
E = 6.3e-5
b = 1300.
jstar = 3.
N0 = 5.24e-3

gm_form = "gk91"
rolloff_option = "no"
A_gk91_unsat = gm_toolbox.vertical_spectrum_Akz(kz, jstar, N, N0, b, gm_form, rolloff_option)
rolloff_option = "yes"
A_gk91_sat = gm_toolbox.vertical_spectrum_Akz(kz, jstar, N, N0, b, gm_form, rolloff_option)

plt.loglog(kz, A_gk91_unsat, label="A_gk91_unsat")
plt.loglog(kz, A_gk91_sat, label="A_gk91_sat")

plt.legend()
plt.show()
```

### Plot GM75, GM76 and GK91 kz-spectra for comparison

```python
plt.loglog(kz, A_gm75_unsat, label="GM75 unsat")
plt.loglog(kz, A_gm75_sat, label="GM75 sat")
plt.loglog(kz, A_gm76_unsat, label="GM76 unsat")
plt.loglog(kz, A_gm76_sat, label="GM76 sat")
plt.loglog(kz, A_gk91_unsat, label="A_gk91_unsat")
plt.loglog(kz, A_gk91_sat, label="A_gk91_sat")

plt.legend()
plt.show()
```
